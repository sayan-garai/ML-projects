```python
import re #regular expression 
import nltk
nltk.download('punkt')        #tokenizer download kar lein hum
from nltk.corpus import stopwords      #importing stopwords
from nltk.stem.snowball import SnowballStemmer      #stemming kar rhe hain idhar
```

```python
sn = SnowballStemmer("english")
stop = set(stopwords.words('english'))
stop
```

```python
def clean_text(sms):   #cleaning of text
    
    
    sms = sms.lower()   #lowercase mein convert kar rhe hain abhi
    sms = re.sub("[^a-z 0-9]", ' ', sms)  #will remove unnecessary colons semicolons apostrophes etc. except numbers 
    sms = nltk.word_tokenize(sms)
    sms = [t for t in sms if len(t)>1]  #length 2 ke neeche ko remove kar dega
    sms = [word for word in sms if word not in stop]  #stop words such as at they etc ko hata dega
    sms = [sn.stem(word) for word in sms if word not in stop]
    sms = ' '.join(sms)
    
    
    return sms
```

```python
clean_text("Badumama MADARCHOD hai EK 1 ka !! he is a bitch") #just checking
```

```python
df['clean_text']= df['text'].apply(clean_text)
df.head()
```

```python
#wordcloud
from wordcloud import WordCloud
```

```python
hamdata = df[df['label'] == 'ham']
hamdata = hamdata['clean_text']   #all the non spams
print(hamdata)
```

```python
def wordCloud(data):
    words = ' '.join(data)
    wc = WordCloud(background_color = "white")
    wc = wc.generate(words)
    
    plt.figure(figsize = (10,8))
    plt.imshow(wc)
    plt.axis("off")
    plt.show()
```

```python
print("Ham data words")
wordCloud(hamdata)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000)
```

```python
X = cv.fit_transform(df['clean_text']).toarray()
```

```python
y = pd.get_dummies(df['label'])   #ham and spam ko bata dega 0s and 1s se
y = y['spam'].values
print(y)
```

```python
#now we'll do the modelling
#now we'll do the modelling
#now we'll do the modelling
#now we'll do the modelling
#now we'll do the modelling
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB 
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train, y_train)  
```

```python
#now the model has been trained, now we'll predict for the test data
#oo yaassssss

y_pred = model.predict(X_test)
```

```python
y_test, y_pred  #comparing them
```

```python
pd.DataFrame({'Actual' : y_test, "Predicted" : y_pred}).head(15)  #first twenty rows comparison
```

```python
#evaluation of the model now, let's goooooooo Garai

print("Accuracy on the Training Data: ")
print((model.score(X_train, y_train))*100)
```

```python
#now the final stage, let's check for the test data Garai!!

print("Accuracy on the test data : ")
print((model.score(X_test , y_test))*100) 


#the accuracy is almost 98% which means it is not overfitting 
```
